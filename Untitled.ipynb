```python
import pandas as pd

df_global = pd.read_csv('Global.csv')
df_munich = pd.read_csv('Munich.csv')
```

```python
df_global.tail(3)
```

```python
df_global.info()
```

```python
df_global_clean = df_global[df_global['year'] <= 2013]
df_global_clean.tail(3)
```

```python
df_global_clean.info()
```

```python
df_munich.tail(3)
```

```python
df_munich.info()
```

```python
df_munich_clean = df_munich[df_munich['year'] >= 1750]
df_munich_clean.head(3)
```

```python
df_munich_clean.info()
```

```python
df_global_ma = df_global_clean
df_global_ma['MA'] = df_global_clean.iloc[:,1].rolling(window = 10).mean()
df_munich_ma = df_munich_clean
df_munich_ma['MA'] = df_munich_clean.iloc[:,3].rolling(window = 10).mean()
```

```python
df_global_ma.head(10)
```

```python
df_munich_ma.head(10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

width = 18
height = 10
plt.figure(figsize=(width, height))

plt.plot(df_global_ma['year'],df_global_ma['MA'],color='red',label='Global_MA')
plt.plot(df_global_ma['year'],df_global_ma['avg_temp'],label='Global')
plt.plot(df_munich_ma['year'],df_munich_ma['MA'],color='magenta',label='Munich_MA')
plt.plot(df_munich_ma['year'],df_munich_ma['avg_temp'],color='green',label='Munich')

plt.grid(True)
plt.legend(loc='lower right')
plt.locator_params(axis='x', nbins=30)
plt.show()
```
